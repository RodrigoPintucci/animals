```python
# plot without first column
df3 = df2.drop(columns=['Idade'])
# green
colors = ['green'] * len(df3.columns)
# paint red if the value is below 50%
for i in range(0, len(df3.columns)):
    if df3.iloc[100][i] < 5:
        colors[i] = 'crimson'

fig = go.Figure(data=[go.Bar(
    x=df3.columns,
    y=df3.iloc[100],
    marker_color=colors # marker color can be a single color value or an iterable
)])
fig.update_layout(title_text='Porcentagem média de confiança')
# order the bars
fig.update_xaxes(categoryorder='total descending')
fig.show()
```
